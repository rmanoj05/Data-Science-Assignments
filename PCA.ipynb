```python
# Let us perform feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
df2 = pd.DataFrame(std.fit_transform(df))
```

```python
df2
```

```python
df.columns
```

```python
df2.rename(columns={0:'Type', 1:'Alcohol', 2:'Malic', 3:'Ash', 4:'Alcalinity', 5:'Magnesium', 6:'Phenols',
       7:'Flavanoids', 8:'Nonflavanoids', 9:'Proanthocyanins', 10:'Color', 11:'Hue',
       12:'Dilution', 13:'Proline'},inplace=True)
```

```python
df2
```

```python
# implementing PCA to reduce dimensionality
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(df2)
```

```python
data_pca = pca.transform(df2)
```

```python
data_pca = pd.DataFrame(data_pca,columns=['P1','P2','P3'])
```

```python
# Principal components of the original dataset
data_pca
```

```python
from sklearn.cluster import KMeans
```

```python
mod1 = KMeans(n_clusters=4)
```

```python
# We will perform clustering on the original dataset
```

```python
lab = mod1.fit_predict(df2)
```

```python
unique = np.unique(lab)
```

```python
x = df2.values
```

```python
centroids = mod1.cluster_centers_ 
```

```python
plt.figure(figsize=(8, 8)) 
for i in unique: 
    plt.scatter(x[lab==i, 0], x[lab==i, 1], label=i) 
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=169, linewidths=3, color='k', zorder=10) 
plt.legend() 
plt.show() 
```

```python
# silhoutte score for clustering done on normal dataframe
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,lab)
```

```python
# Now let us perform clustering on Reduce PCA components
```

```python
mod2 = KMeans(n_clusters=4)
```

```python
label = mod2.fit_predict(data_pca)
```

```python
un = np.unique(label)
```

```python
cent = mod2.cluster_centers_
```

```python
y = data_pca.values
```

```python
plt.figure(figsize=(8, 8)) 
for i in un: 
    plt.scatter(y[label==i, 0], x[label==i, 1], label=i) 
plt.scatter(cent[:, 0], cent[:, 1], marker='x', s=169, linewidths=3, color='k', zorder=10) 
plt.legend() 
plt.show() 
```

```python
# silhouette score for clustering performed on PCA data
```

```python
silhouette_score(y,label)
```
