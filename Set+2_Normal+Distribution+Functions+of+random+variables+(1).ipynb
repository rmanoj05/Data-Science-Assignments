#  Question #2: The current age (in years) of 400 clerical employees at an insurance claims processing center is normally distributed with mean μ = 38 and Standard deviation σ =6. For each statement below, please specify True/False. If false, briefly explain why.



```python
# a)More employees at the processing center are older than 44 than between 38 and 44.
```

```python
#Probability of employees older than 44 P(x>=44)= 1-P(x<=44)
1-stats.norm.cdf(44,38,6)
```

```python
#Probability of employees between 38 and 44 P(x<=38)-P(x<=44)
stats.norm.cdf(44,38,6)-stats.norm.cdf(38,38,6)
```

```python
#Since the probability of an employee to be between ages 38 and 44 is higher therefore the statement is false
```

```python
# b) A training program for employees under the age of 30 at the center would be expected to attract about 36 employees.
```

```python
# Probability of an employee being under 30
stats.norm.cdf(30,38,6)
```

```python
#Number of students attending the training under 30
400 * stats.norm.cdf(30,38,6)
```

```python
#Above statement is true 
```
